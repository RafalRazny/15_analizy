```python
df['DOB'] = pd.to_datetime(df['DOB'],format='%m/%d/%y')
df['DateofHire'] = pd.to_datetime(df['DateofHire'],format='%m/%d/%Y')

df['DateofTermination'] = pd.to_datetime(df['DateofTermination'],format='%m/%d/%y')
```

```python
df[['DateofTermination','DOB','DateofHire']].head()
```

```python
import datetime as dt

def count_seniority(row):

    if pd.isnull(row['DateofTermination']):
        end_date = dt.datetime(2019,9,27)
    else:
        end_date = row['DateofTermination']

    return (end_date - row['DateofHire'])/np.timedelta64(1,'Y')

df['Seniority'] = df.apply(lambda row: count_seniority(row),axis=1)
```

```python
df[['DateofHire','DateofTermination','Seniority']].head()
```

```python
new_table = df.groupby('RecruitmentSource').mean()['Seniority']
new_table.count()
```

```python
new_table_mean = new_table.mean()
new_table_median = new_table.median()
new_table_quantile = new_table.quantile(0.75)
plt.axhline(new_table_mean, color='r', linestyle='--', label='Mean')
plt.axhline(new_table_quantile, color='g', linestyle=':', label='3rd quantile')
plt.axhline(new_table_median, color='b', linestyle='-', label='Median')
plt.legend(['Mean','3rd quantile','Median'])
new_table.plot(kind='bar', figsize=(12,8), width=0.9)
```

Grupa "Married" ocenę 4.0 i 5.0 wystawiła w odpowiednio 31 i 33 przypadkach. Najliczniejsze zatrudnione grupy to "Married" - 123 osoby i "Single" - 137 osób. Można stwierdzić, że w tych dwóch kategoraich osoby z kategorii "Single" lepiej oceniają pracę niż osoby z kategorii "Married".

```python
def count_age(row):
    current_date = dt.datetime(2019,9,27)
    
    if row['DOB'] <= current_date:
        return (current_date - row['DOB'])/np.timedelta64(1,'Y')
    else:
        return 0


df['Age'] = df.apply(lambda row: count_age(row),axis=1)
df['Age']
```

```python
plt.figure(figsize = (8,12))
b=df[df['Age']!=0]
sns.violinplot(y=b['Age'],data=b)
```

### Ad 4
Średnia wieku w firmie wynosi 38 lat, najmłodszy zatrudniony ma 27 lat, najstarszy ponad 50. 3/4 pracowników jest poniżej 43 roku życia. Najwięcej pracowników jest w przedziale wieku 30-35 lat. Najmniej jest osób poniżej 30 i powyżej 45 roku życia. 

```python
def age_category(row):
    if row['Age']<25:
        return '<25'
    elif row['Age']>25 and row['Age']<30:
        return '25-30'
    elif row['Age']>30 and row['Age']<35:
        return '30_35'
    elif row['Age']>35 and row['Age']<40:
        return '35_40'
    elif row['Age']>40 and row['Age']<45:
        return '40_45'
    elif row['Age']>45 and row['Age']<50:
        return '45_50'
    else:
        return '>55'
```

```python
b['AgeCategory'] = b.apply(lambda row: age_category(row),axis=1)
```

```python
b.head(50)
```

```python
z = b.pivot_table(index=b['SpecialProjectsCount'], columns='AgeCategory', values='SpecialProjectsCount', aggfunc='count')
z
```

```python
z.loc["Total"] = z[1:7].sum()
z
```

```python
z = b.pivot_table(index=b['SpecialProjectsCount'], columns='AgeCategory', values='SpecialProjectsCount', aggfunc='count').fillna(0)
z.plot(kind='bar', figsize=(10,8), width=1.0)
plt.xlabel('Number of special projects per person in group category')
plt.ylabel('Number of special projects performed by an employee in last six months in group category')
```
